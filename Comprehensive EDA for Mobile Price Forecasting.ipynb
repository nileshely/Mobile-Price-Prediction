# Mobile Price Prediction Dataset Analysis and Visualization Report

**Introduction:**
The Mobile Price Prediction dataset is a comprehensive collection of mobile phones from various brands, with different specifications and price ranges. This dataset provides a rich source of information for analysis, comparison, or recommendation purposes in the mobile phone industry.

**Dataset Overview:**
The dataset consists of 10 columns, each representing a specific feature of a mobile phone. These features include brand, user ratings, RAM, ROM, mobile size, primary camera resolution, selfie camera resolution, battery power, and price. The dataset contains a total of 836 non-null entries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/mobile-price-prediction-dataset/Mobile Price Prediction Datatset.csv")
```

# Data Visualizations and Insights

```python
# 1. Histogram of Ratings
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='Ratings', bins=30)
plt.title('Distribution of Ratings')
plt.show()
```

A histogram of user ratings shows the frequency of each rating value. Most mobile phones in the dataset have high ratings, indicating good user satisfaction.

```python
# 2. Bar plot of Brands
plt.figure(figsize=(5,80))
sns.countplot(y='Brand me', data=df, order=df['Brand me'].value_counts().index)
plt.title('Number of Mobiles per Brand')
plt.show()
```

A bar plot of brands shows the number of mobiles for each brand in the dataset. This visualization helps identify the brands with the most and least number of mobiles.

```python
# 3. Scatter plot of RAM vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x='RAM', y='Price', data=df)
plt.title('RAM vs Price')
plt.show()
```

A scatter plot of RAM vs Price shows the relationship between the amount of RAM and the price of the mobile phones. There seems to be a positive correlation, indicating that mobile phones with more RAM tend to be more expensive.

```python
# 4. Scatter plot of ROM vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x='ROM', y='Price', data=df)
plt.title('ROM vs Price')
plt.show()
```

A scatter plot of ROM vs Price shows the relationship between the storage capacity and the price of the mobile phones. Similar to RAM, there seems to be a positive correlation, indicating that mobile phones with more storage capacity tend to be more expensive.

```python
# 5. Scatter plot of Battery Power vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x='Battery_Power', y='Price', data=df)
plt.title('Battery Power vs Price')
plt.show()
```

A scatter plot of Battery Power vs Price shows the relationship between the battery capacity and the price of the mobile phones. The correlation here is less clear, suggesting that other factors may play a more significant role in determining the price.

```python
# 6. Box plot of Ratings by Brand
plt.figure(figsize=(5,80))
sns.boxplot(x='Ratings', y='Brand me', data=df)
plt.title('Ratings by Brand')
plt.show()
```

A box plot of Ratings by Brand shows the distribution of ratings for each brand. This visualization helps identify the brands with the highest and lowest user satisfaction.

```python
# 7. Histogram of Mobile Size
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='Mobile_Size', bins=30)
plt.title('Distribution of Mobile Sizes')
plt.show()
```

A histogram of Mobile Sizes shows the frequency of each mobile size. Most mobile phones in the dataset have a size around 5 to 6 inches, which seems to be the preferred size range for users.

```python
# 8. Scatter plot of Primary Camera vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x='Primary_Cam', y='Price', data=df)
plt.title('Primary Camera vs Price')
plt.show()
```

A scatter plot of Primary Camera vs Price shows the relationship between the resolution of the primary camera and the price of the mobile phones. There seems to be a positive correlation, indicating that mobile phones with higher resolution cameras tend to be more expensive.

```python
# 9. Scatter plot of Selfie Camera vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x='Selfi_Cam', y='Price', data=df)
plt.title('Selfie Camera vs Price')
plt.show()
```

A scatter plot of Selfie Camera vs Price shows the relationship between the resolution of the selfie camera and the price of the mobile phones. The correlation here is less clear, suggesting that the resolution of the selfie camera may not be a major factor in determining the price.

```python
# 10. Correlation Heatmap of all features
# Select only numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

A correlation heatmap of all features shows the pairwise correlation between all numeric features in the dataset. This visualization helps identify the features that have a strong positive or negative correlation with the price.
